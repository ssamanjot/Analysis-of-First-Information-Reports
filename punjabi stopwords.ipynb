```python
import pandas as pd
```

```python
data = pd.read_csv("/home/nitpreet/Documents/Sabudh Projects/Fir/FIRDf.csv")
```

```python
f2= open("/home/nitpreet/Documents/Sabudh Projects/stopwords_pun_updated.txt", "r")
stopwords = [x.strip() for x in f2.readlines()]
```

```python
data["cl"] = data["First_Information_Contents"].apply(lambda x: x.replace("।",""))
```

```python
data["cl"]= data["cl"].apply(lambda x : x.replace("#$%-/",""))
```

```python
data["cl"] = data["cl"].apply(lambda x: " ".join([word for word in x.split(" ") if word not in stopwords and len(word)>1]))
```

# Tokenization 

```python
from inltk.inltk import tokenize
```

```python
data["clean"] = data["cl"].apply(lambda x: tokenize(x,language_code='pa'))
```

```python
data["clean"]=data["clean"].apply(lambda x: [word.replace("▁","") for word in x])
```
